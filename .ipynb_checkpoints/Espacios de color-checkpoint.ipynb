## Imagen utilizada

Reutilizo una imagen del radar de apertura sintetica Sentinel correspondiente al embalse de tominé y se recorto a un tamaño de 900 por 1134 píxeles

```python
#Lectura de la imagen
imgSpk3 = cv2.imread('Spk3.png')
tamaño = imgSpk3.shape
tipo = imgSpk3.dtype
plt.imshow(imgSpk3)
print('tamaño y tipo',tamaño,tipo)
```

La imagen se encuentra en el espacio de color  BGR.

La función $ComColor(Imagen,Espacio,EspacioB)$ puede pasar una imagen hasta dos espacios de color diferetes. Sus argumentos son los espacios de color que se desean cambiar y la imagen.

```python
def ComColor(Imagen, EspacioA, EspacioB):
 
    imgEa = cv2.cvtColor(Imagen, EspacioA) 
    
    imgEb = cv2.cvtColor(imgEa, EspacioB)
    
    #Se separan las matrices de la imagen Original
    B,G,R = cv2.split(Imagen)
    
    #Se separan las matrices de la imagen A
    M1_EA,M2_EA,M3_EA = cv2.split(imgEa)
    
    #Se separan las matrices de la imagen B
    M1_EB,M2_EB,M3_EB = cv2.split(imgEb)
    
    #Se imprimen los espacios de color Imagen original, A y B
    figg = plt.figure()

    xe1 = figg.add_subplot(3,3,1)
    xe1.imshow(Imagen)
    xe1.set_title("Imagen BGR")

    xe2 = figg.add_subplot(3,3,5)
    xe2.imshow(imgEa)
    xe2.set_title("Espacio de color A")
    
    xe3 = figg.add_subplot(3,3,9)
    xe3.imshow(imgEb)
    xe3.set_title("Espacio de color B")

    #Se imprimen los compoents de cada espacio de color
    fig = plt.figure()
    
    xf1 = fig.add_subplot(3,9,1)
    xf1.imshow(B)
    xf1.set_title("B")

    xf2 = fig.add_subplot(3,9,11)
    xf2.imshow(G)
    xf2.set_title("G")

    xf3 = fig.add_subplot(3,9,21)
    xf3.imshow(R)
    xf3.set_title("R")

    xf4 = fig.add_subplot(3,9,4)
    xf4.imshow(M1_EA)
    xf4.set_title("Componente a1")

    xf5 = fig.add_subplot(3,9,14)
    xf5.imshow(M2_EA)
    xf5.set_title("Componente a2")

    xf6 = fig.add_subplot(3,9,24)
    xf6.imshow(M3_EA)
    xf6.set_title("Componente a3")

    xf7 = fig.add_subplot(3,9,7)
    xf7.imshow(M1_EB)
    xf7.set_title("Componente b1")

    xf8 = fig.add_subplot(3,9,17)
    xf8.imshow(M2_EB)
    xf8.set_title("Componente b2")

    xf9 = fig.add_subplot(3,9,27)
    xf9.imshow(M3_EB)
    xf9.set_title("Componente b3")
    plt.show()
```

## Conversión de color de RGB a HSV

```python
imgSpk3 = cv2.imread('Spk3.png')
EspacioA = cv2.COLOR_BGR2RGB
EspacioB = cv2.COLOR_RGB2HSV

ComColor(imgSpk3, EspacioA, EspacioB)
```

Espacio A corresponde a la imagen en RGB

Espacio B corresponde a la imagen en HSV

Los componentes a1,a2,a3 corresponden a los componentes HSV, con H = Tono, S = Saturación, V = valor, el cual se puede expresar en %

```python
imgSpk3 = cv2.imread('Spk3.png')
EspacioA = cv2.COLOR_BGR2RGB
EspacioB = cv2.COLOR_RGB2XYZ

ComColor(imgSpk3, EspacioA, EspacioB)
```

Espacio A corresponde a la imagen en RGB

## Conversión de color de LAB a BGR

```python
imgSpk3 = cv2.imread('Spk3.png')
EspacioA = cv2.COLOR_BGR2LAB
EspacioB = cv2.COLOR_LAB2BGR

ComColor(imgSpk3, EspacioA, EspacioB)
```

Espacio A corresponde a la imagen en LAB

Espacio B corresponde a la imagen en BGR

Los componentes a1,a2,a3 corresponden a: L = luminosidad,A = coordenadas rojo/verde (+a indica rojo, -a indica verde),B =  coordenadas amarillo/azul (+b indica amarillo, -b indica azul).

Los componentes b1,b2,b3 corresponden a: B = Azul,G = Verde,R = Rojo.

# 2. Implementar, como una función para OpenCV, las conversiones entre los espacios de color siguientes:

# - RGB a LUX y su inversa.
# - RGB a OHTA y su inversa

```python
def RGB2LUX(Imagen):
    
    imgEa = cv2.cvtColor(Imagen, cv2.COLOR_BGR2RGB)
    
    #Se separan las matrices de la imagen B
    R,G,B = cv2.split(imgEa)

    L = (R+1)**0.3 * (G+1)**0.6 * (B+1)**0.6 - 1
    
    M = 256

    if np.all(R < L):
        U = (M/2)*np.divide((R+1),(L+1))
    else:
        U = M - (M/2)*np.divide((L+1),(R+1))
    
    if np.all(B < L):
        X = (M/2)*np.divide((B+1),(L+1))
    else:
        X = M - (M/2)*np.divide((L+1),(B+1))
    

    return cv2.merge((L,U,X))
    
```

```python
imgSpk3 = cv2.imread('Spk3.png')
img = RGB2LUX(imgSpk3)

R,G,B = cv2.split(imgSpk3)
L,U,X = cv2.split(img)
                  
fig = plt.figure()
xe1 = fig.add_subplot(1,2,1)
xe1.imshow(imgSpk3)
xe1.set_title("Imagen RGB")
xe1.axis("off")

xe2 = fig.add_subplot(1,2,2)
xe2.imshow(img)
xe2.set_title("Imagen LUX")
xe2.axis("off")
    
figg = plt.figure()
    
xf1 = figg.add_subplot(3,6,1)
xf1.imshow(R)
xf1.set_title("R")
xf1.axis("off")

xf2 = figg.add_subplot(3,6,8)
xf2.imshow(G)
xf2.set_title("G")
xf2.axis("off")

xf3 = figg.add_subplot(3,6,15)
xf3.imshow(B)
xf3.set_title("B")
xf3.axis("off")

xf4 = figg.add_subplot(3,6,4)
xf4.imshow( L)
xf4.set_title("L")
xf4.axis("off")

xf5 = figg.add_subplot(3,6,11)
xf5.imshow(U)
xf5.set_title("U")
xf5.axis("off")

xf6 = figg.add_subplot(3,6,18)
xf6.imshow(X)
xf6.set_title("X")
xf6.axis("off")
plt.show()
```

```python
def LUX2RGB(Imagen):
    #Se separan las matrices de la imagen B
    L,U,X = cv2.split(Imagen)
    
    M = 255
    R = L*np.divide(U,M)
    B = L*np.divide(X,M)
    G = np.divide(L,R**0.3 * B**0.1)**1.666666667
   

    return cv2.merge((R,G,B))
        
    
   # G = (np.divide(L+1,(R+1)**0.3 (B+1)**0.6))**1.666666667
```

## RGB a OHTA y su inversa.

```python
def RGB2OHTA(Imaggen):
    imgEEa = cv2.cvtColor(Imaggen, cv2.COLOR_BGR2RGB)
    R,G,B = cv2.split(imgEEa)
    I1 = (1/3) * (R + G + B)
    I2 = (1/2) * (R - B)
    I3 = (1/4) * ((2 * G) - R - B)
    return cv2.merge((I1,I2,I3))
```

```python
imgSpk3 = cv2.imread('Spk3.png')
imgOHTA = RGB2OHTA(imgSpk3)

R,G,B = cv2.split(imgSpk3)
I1,I2,I3 = cv2.split(imgOHTA)
                  
fig = plt.figure()
xe1 = fig.add_subplot(1,2,1)
xe1.imshow(imgSpk3)
xe1.set_title("Imagen RGB")
xe1.axis("off")

xe2 = fig.add_subplot(1,2,2)
xe2.imshow(imgOHTA)
xe2.set_title("Imagen Modelo I1,I2,I3 (OHTA)")
xe2.axis("off")
    
figg = plt.figure()
    
xf1 = figg.add_subplot(3,6,1)
xf1.imshow(R)
xf1.set_title("R")
xf1.axis("off")

xf2 = figg.add_subplot(3,6,8)
xf2.imshow(G)
xf2.set_title("G")
xf2.axis("off")

xf3 = figg.add_subplot(3,6,15)
xf3.imshow(B)
xf3.set_title("B")
xf3.axis("off")

xf4 = figg.add_subplot(3,6,4)
xf4.imshow(I1)
xf4.set_title("I1")
xf4.axis("off")

xf5 = figg.add_subplot(3,6,11)
xf5.imshow(I2)
xf5.set_title("I2")
xf5.axis("off")

xf6 = figg.add_subplot(3,6,18)
xf6.imshow(I3)
xf6.set_title("I3")
xf6.axis("off")
plt.show()
```

```python
def OHTA2RGB(Ima):
    I1,I2,I3 = cv2.split(Ima)
    R = I1 + I2 -((2/3) * I3)
    G = I1 + ((4/3) * I3)
    B = I1 - I2 - ((2/3) * I3) 
    return cv2.merge((R,G,B))
```

```python
imgg = OHTA2RGB(imgOHTA)

I1,I2,I3 = cv2.split(imgOHTA)
R,G,B = cv2.split(imgg)
# Imagen BGR A RGB
imgSpk3 = cv2.imread('Spk3.png')
imgBGR2RGB = cv2.cvtColor(imgSpk3,cv2.COLOR_BGR2RGB)
R2,G2,B2 = cv2.split(imgBGR2RGB)

fig = plt.figure()
xe1 = fig.add_subplot(1,3,1)
xe1.imshow(imgOHTA)
xe1.set_title("Modelo (OHTA)")
xe1.axis("off")

xe2 = fig.add_subplot(1,3,2)
xe2.imshow(imgg)
xe2.set_title("Con. OHTA a RGB")
xe2.axis("off")

xe2 = fig.add_subplot(1,3,3)
xe2.imshow(imgBGR2RGB)
xe2.set_title("BGR a RGB")
xe2.axis("off")
    
figg = plt.figure()
    
xf1 = figg.add_subplot(3,7,1)
xf1.imshow(I1)
xf1.set_title("I1")
xf1.axis("off")

xf2 = figg.add_subplot(3,7,8)
xf2.imshow(I2)
xf2.set_title("I2")
xf2.axis("off")

xf3 = figg.add_subplot(3,7,15)
xf3.imshow(I3)
xf3.set_title("I3")
xf3.axis("off")

xf4 = figg.add_subplot(3,7,4)
xf4.imshow(R)
xf4.set_title("R")
xf4.axis("off")

xf5 = figg.add_subplot(3,7,11)
xf5.imshow(G)
xf5.set_title("G")
xf5.axis("off")

xf6 = figg.add_subplot(3,7,18)
xf6.imshow(B)
xf6.set_title("B")
xf6.axis("off")

xf3 = figg.add_subplot(3,7,15)
xf3.imshow(I3)
xf3.set_title("I3")
xf3.axis("off")

xf4 = figg.add_subplot(3,7,7)
xf4.imshow(R2)
xf4.set_title("R2")
xf4.axis("off")

xf5 = figg.add_subplot(3,7,14)
xf5.imshow(G2)
xf5.set_title("G2")
xf5.axis("off")

xf6 = figg.add_subplot(3,7,21)
xf6.imshow(B2)
xf6.set_title("B2")
xf6.axis("off")
plt.show()
```
